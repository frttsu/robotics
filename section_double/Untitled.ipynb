```python
import sys
sys.path.append('../scripts/')
from mcl_edit import *
```

```python
def trial():
    time_interval = 0.1
    world = World(30, time_interval, debug = False)
    
    m = Map()
    straight = Agent(0.2,0)
    initial_pose = np.array([-4,2,3/2*math.pi]).T
    m.append_landmark(Landmark(initial_pose,agent=straight))
    world.append(m)
    
    initial_pose_1 = np.array([0,0,0]).T
    initial_pose_2 = np.array([-2,5,3 /2*math.pi]).T
    estimator1 = Mcl(m,initial_pose_1,100)
    estimator2 = Mcl(m,initial_pose_2,100)
    a = EstimationAgent(time_interval, 0.2,10.0/180*math.pi, estimator1)
    b = EstimationAgent(time_interval, 0.2,0,estimator2)
    r_1= Robot(initial_pose_1,sensor = Camera(m), agent = a, color = "red")
    r_2 = Robot(initial_pose_2 , sensor = Camera(m),agent = b, color = "blue")
    
    world.append(r_1)
    world.append(r_2)
    
    world.draw()
    
trial()
```
